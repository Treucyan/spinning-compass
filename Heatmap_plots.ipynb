```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["mathtext.fontset"] = "cm" #for latex font
```

#### Fourier Spectrum

```python
lambda_initial, lambda_final = 0.3, 1.5
freq_spectrum = np.loadtxt("raw_data/frequncy spectrum lambda scan.txt")
power_spectrum = np.loadtxt("raw_data/power spectrum lambda scan.txt")

fig, ax = plt.subplots(1, 1, figsize = (10, 4))
diagram = ax.imshow(power_spectrum, origin = "lower", aspect = "auto", extent = [lambda_initial, lambda_final, freq_spectrum[0],\
                                                                        freq_spectrum[-1]], cmap = "inferno")
ax.set_ylim(-3, 3)
diagram.set_clim(0, 0.01)

ax.set_xlabel(r"$\lambda$", fontsize = 25)
ax.set_ylabel(r"$\omega_{r}$", fontsize = 25)


fig.tight_layout()
fig.savefig("image_results/period doubling bifurcations in spin compass.png")
```

#### $B-\omega$ Phase Diagram

```python
B_initial, B_final = 0.2, 1.0
omega_initial, omega_final = 0.5, 1.5
spectral_diagram = np.loadtxt("raw_data/b omega phase diagram data.txt")

fig, ax = plt.subplots(1, 1, figsize = (7, 7))
diagram = ax.imshow(spectral_diagram, cmap = "inferno", extent = [omega_initial, omega_final, B_initial, B_final], origin = "lower", \
          aspect = "auto")

diagram.set_clim(0.2, 0.6)
ax.set_xlabel(r"$\omega / \omega_{0}$", fontsize = 25)
ax.set_ylabel(r"$B / \omega_{0}^{2}$", fontsize = 25)

critical_omega_array = np.linspace(omega_initial, omega_final, 1000)
theory_critical_line = 0.5 * critical_omega_array**2

ax.plot(critical_omega_array, theory_critical_line, "s-", color = "red", markersize = 5)

ax.set_xlim(omega_initial, omega_final)
ax.set_ylim(B_initial, B_final)

fig.tight_layout()
fig.savefig("image_results/b omega phase diagram.png")
```
