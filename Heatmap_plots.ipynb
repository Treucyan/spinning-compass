```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["mathtext.fontset"] = "cm" #for latex font
```

#### Fourier Spectrum

```python
lambda_initial, lambda_final = 0.3, 1.5
freq_spectrum = np.loadtxt("raw_data/frequncy spectrum lambda scan.txt")
power_spectrum = np.loadtxt("raw_data/power spectrum lambda scan.txt")

fig, ax = plt.subplots(1, 1, figsize = (10, 4))
diagram = ax.imshow(power_spectrum, origin = "lower", aspect = "auto", extent = [lambda_initial, lambda_final, freq_spectrum[0],\
                                                                        freq_spectrum[-1]], cmap = "inferno")
ax.set_ylim(-2, 2)
diagram.set_clim(0, 0.01)

ax.set_xlabel(r"$\lambda$", fontsize = 25)
ax.set_ylabel(r"$\omega_{r}$", fontsize = 25)

fig.tight_layout()
fig.savefig("image_results/period doubling bifurcations in spin compass.png")
```
